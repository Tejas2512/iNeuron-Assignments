# Problem 1

We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
month = np.array([x for x in range(1,13)])
```

```python
plt.figure(figsize=(10,7))
plt.plot(month,Max,'bo',month,Max,'k',color='red')
plt.plot(month,Min,'bo',month,Min,'k')
plt.xlabel('Month')
plt.ylabel('Temparature')
plt.grid()
```

# Problem 2

data to use:
url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv

```python
df = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
df.head()
```

```python
df.sex.fillna(value='male',inplace=True)
df.age.fillna(value=df.age.mean(),inplace=True)
df.fare.fillna(value=df.fare.mean(),inplace=True)
percentage = (df.sex.value_counts()/len(df))*100
```

1. Create a pie chart presenting the male/female proportion

```python
plt.pie(percentage,labels=['male','female'])
plt.title('presenting the male/female proportion')
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
cols = {'male': 'orange', 'female': 'blue'}
plt.figure(figsize=(7,5))
plt.scatter(y='fare',x='age',data=df,c=df.sex.map(cols))
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid()
```
