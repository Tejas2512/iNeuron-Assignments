# Problem

In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.


```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from mpl_toolkits import mplot3d
```

```python
df = datasets.load_iris()
```

```python
df1 = pd.DataFrame(df.data)
df1.rename(columns={0:"sepal_length",1:"sepal_width",2:"petal_length",3:"petal_width"},inplace=True)
df1 = pd.concat([df1,pd.DataFrame(df.target)],axis=1)
df1.rename(columns={0:"Target"},inplace=True)
```

```python
color = {0:'red',1:'yellow',2:'green'}
plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.scatter3D(df1["sepal_length"],df1["sepal_width"],df1["petal_length"],c=df1.Target.map(color))   
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
ax.set_zlabel("petal_length")
ax.set_title("3d plot")
```
